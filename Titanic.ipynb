```python
# Mostrando Gráficos com Python no Notebook
%matplotlib inline
```

```python
# Usando Matplotlib para desenhar um histogram para a coluna Age(idade)
plt.hist(df['Age'],bins=8)
```

```python
# Usando Matplotlib para desenhar um histogram para a coluna Sex
plt.hist(df['Sex'],bins=8)
```

```python
# Usando Matplotlib para desenhar um histogram para a coluna Survived em que 0 = não sobreviveu e 1 = sobreviveu
plt.hist(df['Survived'],bins=8)
```

```python
# Usando o seaborn package para desenhar o histograma das idades
sns.distplot(df['Age']);
```

```python
# O uso regular do matplotlib para mostrar o "Gráfico de Barras" barplot e a quantidade de sobreviventes de cada sexo
df.groupby(['Sex'])['Survived'].mean().plot(kind='pie')
```

```python
# O uso do seaborn para mostrar o Gráfico de Barras e a quantidade de sobreviventes de cada sexo
sns.set_style("whitegrid")
ax = sns.barplot(x='Sex',y ='Survived', data=df, estimator=len)
```

```python
# Dividindo o gráfico em 02 grupos:
ax = sns.barplot(x='Sex',y ='Age', hue='Survived', data=df, estimator=len)
```

```python
#Violinplot
sns.violinplot(x = "Age", data=df)
```

```python
#Scatterplot in seaborn
sns.jointplot(x='Sex', y='Survived', data=df)
```

```python
#Se estivermos em gráficos para representar a regressão linear para 2 variáveis numéricas - regplot
sns.regplot(x='Age', y='Survived', data=df)
```

```python
# box plot - é um tipo de Gráfico muito utilizado para detectar Out-lier
sns.boxplot(x='Sex',y='Age', hue='Survived', data=df)
```

```python
# side-by-side box plot - com parâmetros de sexo com as idades aplicadas aos sobreviventes
sns.boxplot(x='Sex',y='Age', data=df, hue='Survived')
```
